```python
def cleanhtml(raw_html):
    raw_html = str(raw_html) #incase its not a string xd
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', raw_html)
    return " ".join(cleantext.split())
# print(cleanhtml(str(review_votes[0])))
```

```python
def getnum(s):
    '''
    parses for numbers and decimals, note returns STRING
    '''
    numbers = re.compile('\d+(?:\.\d+)?')
    return numbers.findall(s)[0]
```

```python
def build_prof_db():
    '''
    Updates/Builds Database
    TODO if there's a load more button, either change the html or somehow click the button
    '''
    
    adds = 0
    prof_id = 2
    misses = 0
    while True:
        if misses==20:
            return
        if adds%20==0:
            print("{} adds".format(adds))
#         if adds>=5:
#             break
        try:
    #         try: #how to see if invalid query
            url = requests.get("http://www.ratemyprofessors.com/ShowRatings.jsp?tid={}&showMyProfs=true".format(prof_id))
            source = BeautifulSoup(url.text,'html.parser')
            #Professor Information
            name = cleanhtml(source.find_all('h1',class_='profname'))
            school = source.find('a',class_='school').get_text()
            title = cleanhtml(source.find('div',class_='result-title').get_text())
            department = title[title.index("the")+4:title.index("department")]
            tags = tag_feed.find_all('span',class_='tag-box-choosetags')
            tags = [t.get_text().strip() for t in tags]
            tag_counts = [getnum(t) for t in tags]
            tag_names = [tags[i][:tags[i].index(tag_counts[i])-2] for i in range(len(tags))]
            tag_dict = {}
            for i in range(len(tag_counts)):
                tag_dict[tag_names[i]] = tag_counts[i]

            #Reviews
            review_feed = source.find('table',class_="tftable")
            dates = review_feed.find_all('div',class_='date')
            num_reviews = len(dates)
            dates = [d.get_text() for d in dates]
            reviews = review_feed.find_all('div',class_='descriptor-container') # overall, difficult, overall, difficult,
            ratings,difficultys = [],[]
            for i in range(len(reviews)):
                if i%2==1:
                    difficultys.append(getnum(reviews[i].get_text()))
                else:
                    ratings.append(getnum(reviews[i].get_text()))

            resp = review_feed.find_all('span',class_='response') #course number, for credit, attendance, textbook used, would take again, grade received
            comment_body = review_feed.find_all('p',class_='commentsParagraph')
            comments = [cleanhtml(c.get_text()) for c in comment_body]
            review_votes = review_feed.find_all('span',class_='count') #[upvote1,downvote1,upvote2,...]
            upvotes,downvotes = [],[]
            for i in range(len(review_votes)):
                if i%2==1:
                    downvotes.append(getnum(review_votes[i].get_text()))
                else:
                    upvotes.append(getnum(review_votes[i].get_text()))
            review_array = []
            for i in range(num_reviews):
                review_array.append({"date":dates[i],"rating":ratings[i],"difficulty":difficultys[i],"comment":comments[i],"upvotes":upvotes[i],"downvotes":downvotes[i]})            
            professors[prof_id] = {"name":name,"school":school,"title":title,"department":department,"tags":tag_dict,"reviews":review_array}
            adds+=1
            misses=0
        except:
            misses+=1
        prof_id+=1
        print("{} misses".format(misses))
        print("{} adds".format(adds))

build_prof_db()        
# print(professors)
```
